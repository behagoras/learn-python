## 2. Match Statement (Switch)

Introduced in Python 3.10, the `match` statement evaluates an expression and compares its value against various patterns (`case`).

### Key Concepts:
* **Clean Syntax:** Replaces messy chains of `if/elif/elif/else`.
* **Value Matching:** Ideal for checking specific values (like status codes, days of the week, or commands).
* **The Wildcard (`_`)**: Acts like an `else`. It captures anything that didn't match the specific cases above it.

## 4. For Loop & Range

The `for` loop acts as an **iterator**. It processes elements of a sequence (like a list, string, or range) one by one.

### The `range()` function:
It generates a sequence of numbers. Syntax: `range(start, stop, step)`.
* **Start:** Inclusive (default 0).
* **Stop:** **Exclusive** (it stops *before* this number).
* **Step:** The increment (default 1).

Example: `range(0, 10, 2)` gives `0, 2, 4, 6, 8`.
